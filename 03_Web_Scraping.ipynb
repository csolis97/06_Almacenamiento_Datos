```python
# Chequeo de status de conexión a la pagina
url = 'https://es.wikipedia.org/wiki/Regiones_de_Chile'
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'}

response = requests.get(url, headers=headers)
response.status_code
```

```python
# Contenido HTML de la pagina
bs = BeautifulSoup(response.content)
bs
```

```python
# Titulo de la pagina
print(bs.title.text)
```

```python
print(bs.find_all('a'))
```

```python
# Encontrar todas las tablas en la página
soup = BeautifulSoup(response.text, 'html.parser')
tablas = soup.find_all('table', {'class': 'wikitable'})

# Elegir la tabla que necesitas (por índice, según las tablas encontradas)
tabla_regiones = tablas[1] 
# Procesar filas de la tabla y extraer los datos
filas = tabla_regiones.find_all('tr')
datos = []

# Extraer cabeceras
headers = [th.text.strip() for th in filas[0].find_all('th')]

# Extraer datos de cada fila
for fila in filas[1:]:
    columnas = fila.find_all(['td', 'th'])
    datos.append([col.text.strip() for col in columnas])

# Convertir a DataFrame
df_regiones = pd.DataFrame(datos, columns=headers)

# Mostrar las primeras filas
print(df_regiones.head())
```

Se obtiene el mismo resultado, pero con mucho más ezfuerzo computacionaol

```python
with open('books.html','r') as html_file:
    content = html_file.read()

print(content)
```

```python
# Utilizamos BeatifulSoup para parser el contenido del archivo local HTML
lxml_reader = BeautifulSoup(content, 'lxml')
```

```python
autores = lxml_reader.find_all('h5')

for autor in autores:
    print (autor.text)
```

```python
libros = lxml_reader.find_all('a')

for libro in libros:
    print(libro.text)
```

```python
autores_obras = lxml_reader.find_all('div', class_ = 'card')

for autor_obra in autores_obras:
    autores = autor_obra.h5.text
    obras = autor_obra.a.text
    print(autores, obras)
```
