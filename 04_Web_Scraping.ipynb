```python
# Importamos las librerías necesarias
import pandas as pd
from bs4 import BeautifulSoup
import requests
import lxml
```

```python
url = 'https://www.chiletrabajos.cl/encuentra-un-empleo?action=search&order_by=&ord=&within=25&2=data+scientist&filterSearch=Buscar'
response = requests.get(url)
response.status_code
```

```python
soup = BeautifulSoup(response.text,'lxml')
soup
```

```python
ofertas = soup.find_all('div', {'class': 'job-item with-thumb destacado no-hover'})
```

```python
for oferta in ofertas:
    cargo = oferta.find('a', class_='font-weight-bold').text
    descripción = oferta.find('p', class_='description').text.strip()
    print(f'{cargo}, \n {descripción}')
```

```python
# Definimos la listas iniciales
cargos = []  
empresas = []
descripciones = []
links = []

# Iteramos sobre cada job-item para extraer los datos
for job_item in ofertas:
    # Extraemos el cargo
    cargo = job_item.find('h2', class_='title')
    if cargo:
        cargos.append(cargo.text.strip())
    else:
        cargos.append(None)

    # Extraemos el nombre de la empresa
    empresa = job_item.find('h3', class_='meta')
    if empresa:
        # Separamos el nombre de la empresa y la ubicación, usando la coma como delimitador
        empresa_texto = empresa.text.strip().split(',')[0]  # Tomamos solo la primera parte antes de la coma
        empresas.append(empresa_texto)
    else:
        empresas.append(None)

    # Extraemos la descripción (opcional, ya que no siempre está presente)
    descripcion = job_item.find('p', class_='description')
    if descripcion:
        descripciones.append(descripcion.text.strip())
    else:
        descripciones.append(None)

    # Extraemos el link de la oferta
    link = job_item.find('a', href=True)
    if link:
        links.append(link['href'])
    else:
        links.append(None)

# Creamos un DataFrame con los datos obtenidos
df = pd.DataFrame({
    'cargo': cargos,
    'empresa': empresas,
    'descripcion': descripciones,
    'link': links
})

# Muestra el DataFrame
df
```
